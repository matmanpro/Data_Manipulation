## Birleştirme (Join) İşlemleri

```python
import numpy as np
import pandas as pd

m = np.random.randint(1,30, size = (5,3))

df1= pd.DataFrame(m, columns = ["var1","var2","var3"])
df1
```

```python
df2 = df1 +99
df2
```

```python
#  hata ---- indeksler 0-4 0-4 olacak
pd.concat([df1,df2])
```

```python
# yukarıdaki hatayı gidermek için : 

pd.concat([df1,df2], ignore_index= True)
```

```python
df2.columns = ["var1","var2","deg3"]
df2
```

```python
df1
```

```python
pd.concat([df1,df2])
```

```python
# kesişimlere göre birleştir
pd.concat([df1,df2], join = "inner", ignore_index =True)
```

```python
pd.concat([df2,df1.reindex(columns=df2.columns)])
```
